### Modules Required:

1. **Numpy:** NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.


2. **Matplotlib:** Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy. 


3. **cv2:** OpenCV-Python is a library of Python bindings designed to solve computer vision problems.

```python
## importing required modules
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
import cv2
```

```python
## module versions
print('Numpy version: {}'.format(np.__version__))
print('Matplotlib version: {}'.format(mlp.__version__))
print('cv2 version: {}'.format(cv2.__version__))
```

**1. Reading in a image:-**

```python
## reading in the image
img = cv2.imread('the_ball.jpg', cv2.IMREAD_GRAYSCALE)

## reading image as RGB or in cv2 it's BGR
## img = cv2.imread('the_ball.jpg')

## showing the image
cv2.imshow('image', img)
cv2.waitKey(0)   ## wait for a key to close the window
cv2.destroyAllWindows()
```

**2. Displaying images using matplotlib:**

```python
## diplaying the image
plt.imshow(img, cmap='gray', interpolation='bicubic')

## cmap set to gray because matplotlib uses RGB format while cv2 uses BGR, so explicitly given the color map.
## 'bicubic interpolation' is an interpolation technique for sharpening enlargements of digital images.
```

**3. Saving the image:**

```python
## saving the image using imwrite method
cv2.imwrite('the_ball_grey.jpg', img)
```
