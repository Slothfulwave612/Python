```python
## importing required modules
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

% matplotlib inline
```

```python
## modules version
print('Pandas Version: {}'.format(pd.__version__))
print('Matplotlib Version: {}'.format(matplotlib.__version__))
```

```python
## data for our plots
data = pd.read_csv('data.csv')

## ages
ages = data['Age']

## dev salaries
dev_salary = data['All_Devs']
py_salary = data['Python']
js_salary = data['JavaScript']

## median salary
overall_median = 57287
```

**1. Creating a simple line plot:-**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting the line
plt.plot(ages, dev_salary, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salary, label='Python')

## adding labels and title
plt.title('Median Salary (USD) by Age')
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')

## adding legends
plt.legend()
```

**2. Filling the are below the Py Dev salary:-**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting the line plots
plt.plot(ages, dev_salary, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salary, label='Python')

## filling below our python salaries
plt.fill_between(x=ages, y1=py_salary, alpha=0.25)

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')

## adding legend
plt.legend()
```

**3. Filling between Py Dev Salaries and Dev Salaries:**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting line plots
plt.plot(ages, dev_salary, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salary, label='Python')

## filling the area between the py_salary and dev_salary
plt.fill_between(x=ages, y1=dev_salary, y2=py_salary, alpha=0.25)

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')

## adding legend
plt.legend()
```

**4. Filling between Py Dev salaries and Median Salaries:**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting line plots
plt.plot(ages, dev_salary, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salary, label='Python')

## filling between py_dev salary and overall median
plt.fill_between(x=ages, y1=py_salary, y2=overall_median, alpha=0.25)

## adding labels and titles
plt.xlabel('Ages')
plt.ylabel('Medain Salary (USD)')
plt.title('Medain Salary (USD) by Age')

## adding legend
plt.legend()
```

**5. Conditional Filling:**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting line plots
plt.plot(ages, dev_salary, color='#444444', linestyle='--', label='All Devs')
plt.plot(ages, py_salary, label='Python')

## filling between using condition
plt.fill_between(x=ages, y1=py_salary, y2=overall_median, 
                where=(py_salary > overall_median), interpolate=True, alpha=0.25, label='Above Avg')       
plt.fill_between(x=ages, y1=py_salary, y2=overall_median,
                where=(py_salary <= overall_median), color='red', interpolate=True, alpha=0.25, label='Below Avg')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by Age')

## adding legends
plt.legend()
```
