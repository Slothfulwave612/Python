### Module Required:
1. Matplotlib:- Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

```python
## importing matplotlib module
import matplotlib.pyplot as plt
% matplotlib inline
```

**1. Line Plot:-**

```python
## data for our line plot
ages = list(range(25, 36))

dev_salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]

py_dev_salary = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
## plotting line plot
plt.plot(ages, dev_salary)
```

**2. Adding Labels and Titles:-**

```python
## label for x-coordinate, y-coordinate and adding title to the plot
plt.plot(ages, dev_salary)
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')
```

**3. Plotting multiple lines on the plot:-**

```python
## plotting median salary for all devs and python devs
plt.plot(ages, dev_salary)
plt.plot(ages, py_dev_salary)

## adding labels and all
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')
```

**4. Adding Legends:-**

```python
## plotting meadian salary for all devs and python devs
plt.plot(ages, dev_salary)
plt.plot(ages, py_dev_salary)

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legend
plt.legend(['All Devs', 'Py Dev'])
```

**5. Another Way To Add Legends:-**

```python
## plotting median salary for all devs and python devs
plt.plot(ages, dev_salary, label='All Dev')
plt.plot(ages, py_dev_salary, label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legends
plt.legend()
```

**6. Adding Colors:-**

```python
## plotting median salary for all devs and python devs
plt.plot(ages, dev_salary, 'k--', label='All Dev')
plt.plot(ages, py_dev_salary, 'b-o', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legends
plt.legend()
```

**7. Another Way:-**

```python
## plotting median salary for all devs and python devs
plt.plot(ages, dev_salary, color='k', linestyle='--', label='All Dev')
plt.plot(ages, py_dev_salary, color='b', linestyle='dotted', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legends
plt.legend()
```

**8. Marker Size:-**

```python
## plotting a line plot
## using hex color code 
plt.plot(ages, dev_salary, color='#554251', linestyle='--', marker='o', label='All Dev')
plt.plot(ages, py_dev_salary, color='#800000', linestyle='--', marker='.', label='Py Dev')

## adding label and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legends
plt.legend()
```

**9. Changing Width of a Line:-**

```python
## plotting line plot
plt.plot(ages, dev_salary, color='#00FFFF', linewidth=3,label='All Dev')
plt.plot(ages, py_dev_salary, color='#800000', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legends
plt.legend()

## you can use: plt.tight_layout() for padding
```

**10. Adding grid in plot:-**

```python
## plotting line plots
plt.plot(ages, dev_salary, color='#008000', label='All Dev')
plt.plot(ages, py_dev_salary, color='#800000', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding grid to the plot
plt.grid(True)

## adding legends
plt.legend()
```

**11. To see all the styles available to us on matplotlib:-**

```python
plt.style.available
```

**12. Using a style:-**

```python
## using a style for our plot
plt.style.use('fivethirtyeight')

## plotting line plots
plt.plot(ages, dev_salary, color='#008000', label='All Dev')
plt.plot(ages, py_dev_salary, color='#800000', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary (USD)')
plt.title('Median Salary (USD) by age')

## adding legends
plt.legend()
```

**13. Saving Plots:-**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting line plots
plt.plot(ages, dev_salary, color='#008000', label='All Dev')
plt.plot(ages, py_dev_salary, color='#800000', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by age')

## adding legends
plt.legend()

## saving figure
plt.savefig('line_plot')
```
