```python
## importing required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

% matplotlib inline
```

```python
## data for our plot
ages = np.arange(25,36,1)

dev_salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]

py_dev_salary = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]

js_dev_y = [37810, 43515, 46823, 49293, 53437,
             56373, 62375, 66674, 68745, 68746, 74583]
```

**1. Creating Bar Plots:-**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting a simple bar plot
plt.bar(ages, py_dev_salary, color='#008000', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by age')

## adding legend
plt.legend()
```

**2. Creating Multiple Bars:-**

```python
## using seaborn style
plt.style.use('seaborn')

## defining some parameters that will help us in plotting multiple bars
x_indexes = np.arange(len(ages))
width = 0.25

## plotting bars
plt.bar(ages - width, dev_salary, width=width, label='All Dev')
plt.bar(ages, py_dev_salary, width=width, label='Py Dev')
plt.bar(ages + width, dev_salary, width=width, label='JS Dev')

## adding labels and titles
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by age')

## for showing each and every age
plt.xticks(ticks=ages, labels=ages)

## adding legends
plt.legend()
```

**3. Making Horizontal Bars:-**

```python
## reading in the data from csv file: data.csv
data = pd.read_csv('data.csv')
data.head()
```

```python
## storing LanguagesWorkedwWith
langs = data['LanguagesWorkedWith']

## counting every occurence of each language
lang_counter = Counter()

for lang in langs:
    lang_counter.update(lang.split(';'))

lang_counter
```

```python
## most common 15 languages
lang_counter.most_common(15)
```

```python
## making list of most common 15 languages and their count
lang, counter = [], []

for item in lang_counter.most_common(15):
    lang.append(item[0])
    counter.append(item[1])

print(lang, end='\n\n')
print(counter)
```

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## making a horizontal bar plot
plt.barh(lang, counter)
```

```python
## reverse the value so that the most common language will be shown at the top
lang.reverse()
counter.reverse()
```

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## horizontal bar
plt.barh(lang, counter)
```
