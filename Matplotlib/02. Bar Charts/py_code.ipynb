### Module Required:
1. **Numpy** :- NumPy is a library for the Python programming language, adding support for large, multi-dimensional arrays and matrices, along with a large collection of high-level mathematical functions to operate on these arrays.


2. **Pandas**:- In computer programming, pandas is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series.


3. **Matplotlib**:- Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

```python
## importing required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

% matplotlib inline
```

```python
## data for our plot
ages = np.arange(25,36,1)

dev_salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]

py_dev_salary = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]

js_dev_y = [37810, 43515, 46823, 49293, 53437,
             56373, 62375, 66674, 68745, 68746, 74583]
```

**1. Creating Bar Plots:-**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting a simple bar plot
plt.bar(ages, py_dev_salary, color='#008000', label='Py Dev')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by age')

## adding legend
plt.legend()
```

**2. Creating Multiple Bars:-**

```python
## using seaborn style
plt.style.use('seaborn')

## defining some parameters that will help us in plotting multiple bars
x_indexes = np.arange(len(ages))
width = 0.25

## plotting bars
plt.bar(ages - width, dev_salary, width=width, label='All Dev')
plt.bar(ages, py_dev_salary, width=width, label='Py Dev')
plt.bar(ages + width, dev_salary, width=width, label='JS Dev')

## adding labels and titles
plt.xlabel('Ages')
plt.ylabel('Median Salary(USD)')
plt.title('Median Salary(USD) by age')

## for showing each and every age
plt.xticks(ticks=ages, labels=ages)

## adding legends
plt.legend()
```
