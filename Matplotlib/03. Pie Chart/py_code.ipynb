### Module Required:
1. **Matplotlib**:- Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.

```python
## importing required modules
import matplotlib.pyplot as plt
import matplotlib

% matplotlib inline
```

```python
## modules version
matplotlib.__version__
```

```python
## data for plotting
slices_1 = [60, 40]

slices_2 = [120, 80, 30, 20]

slices_3 = [10124, 52476, 78452, 83471, 72814]

labels_1 = ['Sixty', 'Forty']

labels_2 = ['Parameter_1', 'Parameter_2', 'Parameter_3', 'Parameter_4']

labels_3 = ['Parameter_1', 'Parameter_2', 'Parameter_3', 'Parameter_4', 'Parameter_5']

## color options
colors = ['#008fd5', '#fc4f30', '#e5ae37', '#6d90df']

colors_2 = ['#008fd5', '#fc4f30', '#e5ae37', '#6d90df', '#228B22']
```

**1. A simple pie chart:-**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## plotting a pie chart
plt.pie(slices_1)
```

**2. Adding labels:-**

```python
## using seaborn style
plt.style.use('seaborn')

## plotting pie chart with labels
plt.pie(slices_1, labels=labels_1)
```

**3. Edge Color:-**

```python
## using seaborn style
plt.style.use('seaborn')

## plotting pie chart
plt.pie(slices_1, labels=labels_1, wedgeprops={'linewidth': 1.3, 'edgecolor': 'black'})
```

**4. Changing colors of the slices:-**

```python
## using ggplot style
plt.style.use('ggplot')

## plotting pie
plt.pie(slices_2, labels=labels_2, colors=colors, wedgeprops={'linewidth': 1.3, 'edgecolor': 'black'})
```

**5. Emphasizing a particular slice in a pie chart:-**

```python
## using ggplot style
plt.style.use('ggplot')

## using explode
explode = [0, 0, 0, 0.09, 0]     ## to emphasize on Parameter_4 by 9% of the radius of the circle

## plotting pie
plt.pie(slices_3, labels=labels_3, colors=colors_2, explode=explode, wedgeprops={'linewidth': 1.15, 'edgecolor': 'black'})
```

**6. Shadow pie chart:-**

```python
## using seaborn style
plt.style.use('seaborn')

## plotting pie
plt.pie(slices_3, colors=colors_2, labels=labels_3, shadow=True, explode=explode, 
        wedgeprops={'linewidth': 1.15, 'edgecolor': 'black'})
```

**7. Rotate pie chart:-**

```python
## using seaborn style
plt.style.use('seaborn')

## plotting pie
plt.pie(slices_3, labels=labels_3, colors=colors_2, shadow=True, wedgeprops={'linewidth': 1.15, 'edgecolor': 'black'},
       explode=explode, startangle=180)
```

**8. Adding percentage as a label:-**

```python
## using seaborn style
plt.style.use('seaborn')

## plotting pie
plt.pie(slices_3, labels=labels_3, colors=colors_2, shadow=True, startangle=180, 
       wedgeprops={'linewidth': 1.15, 'edgecolor': 'black'}, explode=explode, autopct='%1.1f%%')
```
