### Module Required:
1. **Pandas**:- In computer programming, pandas is a software library written for the Python programming language for data manipulation and analysis. In particular, it offers data structures and operations for manipulating numerical tables and time series.

```python
## importing required modules
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib

% matplotlib inline
```

```python
## modules version
print('Pandas Version: {}'.format(pd.__version__))
print('Matplotlib Version: {}'.format(matplotlib.__version__))
```

```python
## data for our plots
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

## loading from data.csv
data = pd.read_csv('data.csv')
ids = data['Responder_id']
ages_ = data['Age']
median_age = 29
```

**1. Creating a simple histogram:**

```python
## using seaborn style
plt.style.use('seaborn')

## plotting a histogram
plt.hist(ages, bins=5, edgecolor='black')

## adding labels and titles
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Histogram')
```

**2. Using our own bins:**

```python
## using ggplot style
plt.style.use('ggplot')

## using our bins
bins = [10, 20, 30, 40, 50, 60]

## plotting histogram
plt.hist(ages, bins=bins, edgecolor='black', color='#9999FF')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Histogram')
```

**3. Histogram should not plot ages between 10 to 20:**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## bins
bins = [20, 30, 40, 50, 60]

## plotting histogram
plt.hist(ages, bins=bins, color='#9999FF', edgecolor='#444444')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Histogram')
```

**4. Plotting a histogram for data.csv:**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## bins
bins = [10, 20, 30, 40, 50, 60, 70, 80]

## plotting a histogram
plt.hist(ages_, bins=bins, color='#9999FF', edgecolor='#444444')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age of Respondents')
```

**5. The log parameter:**

```python
## using fivethirtyeight style
plt.style.use('fivethirtyeight')

## bins
bins = [10, 20, 30, 40, 50, 60, 70, 80]

## plotting histogram
plt.hist(ages_, bins=bins, color='#9999FF', edgecolor='#444444', log=True)

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age of Respondents')
```

**6. Plotting Median Age:**

```python
## using fivethirtyeight style
plt.style .use('fivethirtyeight')

## bins
bins = [10, 20, 30, 40, 50, 60, 70]

## plotting histogram
plt.hist(ages_, bins=bins, color='#9999FF', edgecolor='#444444')

## plotting median age
plt.axvline(median_age, color='#bc4f87', label='median age')

## adding labels and title
plt.xlabel('Ages')
plt.ylabel('Count')
plt.title('Age of Respondents')

## showing legend
plt.legend()
```
